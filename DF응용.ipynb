함수 매핑

```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df['ten'] = 10
print(df.head())
```

```python
def add_10(n):
  return n+10
def add_two_obj(a,b):
  return a+b

sr1 = df['age'].apply(add_10)
print(sr1.head())
sr2 = df['age'].apply(add_two_obj, b=10)
print(sr2.head())

df_map = df.applymap(add_10)
print(df_map.head())
```

```python
def missing_value(series):
  return series.isnull()

result = df.apply(missing_value)
print(result.head())
```

```python
def min_max(x):
  return x.max() - x.min()

result = df.apply(min_max)
print(result.head())
```

```python
#각 행에 함수 매핑 : axis = 1
df['add'] = df.apply(lambda x: add_two_obj(x['age'],x['ten']), axis=1)
print(df.head())
```

```python
#데이터프레임 객체에 함수 매핑: pipe()
def missing_count(x):
  return missing_value(x).sum()
def total_number_missing(x):
  return missing_count(x).sum()

result_df = df.pipe(missing_value)
print(result_df.head())

result_series = df.pipe(missing_count)
print(result_series)

result_value = df.pipe(total_number_missing)
print(result_value)
```

데이터프레임 합치기

```python
#axis=0이면 위아래, 1이면 양옆으로~
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
print(df1)
print(df2)
result = pd.concat([df1,df2])
print(result)
```

```python
result2 = pd.concat([df1,df2], ignore_index = True)#기존 index 유지안됨
print(result2)
```

```python
result3 = pd.concat([df1,df2], axis=1)
print(result3)
```

```python
result3_in = pd.concat([df1,df2], axis=1, join='inner')#중복되는거만
print(result3_in)
```

```python
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name='f', index=[3, 4, 5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name='g')

result4 = pd.concat([df1,sr1],axis=1)
print(result4)
```

```python
#데이터 합치기: merge()
df1 = pd.read_excel('./stock price.xlsx')
df2 = pd.read_excel('./stock valuation.xlsx')

print(df1)
print(df2)
```

```python
merge_inner = pd.merge(df1,df2)
print(merge_inner)
```

```python
merge_outer = pd.merge(df1,df2, how='outer',on='id')#id열을 기준으로 모든 종목의 데이터 통합
print(merge_outer)
```

```python
merge_left = pd.merge(df1,df2,how='left',left_on='stock_name',right_on='name')
print(merge_left)
```

```python
price = df1[df1.price < 50000]
print(price)
print(pd.merge(price,df2))
```

```python
#행 인덱스 기준으로 병합: join
df1 = pd.read_excel('./stock price.xlsx', index_col='id')
df2 = pd.read_excel('./stock valuation.xlsx', index_col='id')

df3 = df1.join(df2)
print(df3)
df4 = df1.join(df2, how='inner')#공통으로 존재하는 행 인덱스 기준으로 병합
print(df4)
```

그룹 연산

```python
#분할
df = titanic.loc[:,['age','sex','class','fare','survived']]
print('승객 수:', len(df))

grouped = df.groupby(['class'])
for key, group in grouped:
  print('* key: ',key)
  print('* number: ', len(group))
  print(group.head())
  print('\n')
```

```python
average = grouped.mean()
print(average)
```

```python
group3 = grouped.get_group('Third')
print(group3.head())
```

```python
grouped_two = df.groupby(['class','sex'])
for key, group in grouped_two:
  print('* key: ',key)
  print('* number: ', len(group))
  print(group.head())
  print('\n')
```

```python
average_two = grouped_two.mean()
print(average_two)
```

```python
group3f = grouped_two.get_group(('Third','female'))
print(group3f.head())
```

```python
#적용-결합
grouped = df.groupby(['class'])

std_all = grouped.std()
print(std_all)
std_fare = grouped.fare.std()
print(std_fare)
```

```python
#집계 연산 함수 적용: agg()
def mean_max(x):
  return x.max() - x.min()
agg_minmax = grouped.agg(min_max)
print(agg_minmax.head())
```

```python
#모든 열에 같은 함수 매핑
agg_all = grouped.agg(['min','max'])
print(agg_all.head())
#각 열에 각각 함수 매핑
agg_sep = grouped.agg({'fare':['min','max'],'age':'mean'})
print(agg_sep.head())
```

```python
age_mean = grouped.age.mean()
age_std = grouped.age.std()

for key, group in grouped.age:
  group_zscore = (group-age_mean.loc[key])/age_std.loc[key]
  print('* key: ',key)
  print(group_zscore.head())
  print('\n')
```

```python
def z_score(x):
  return (x-x.mean())/x.std()

age_zscore = grouped.age.transform(z_score)
print(age_zscore.head())
```

```python
#그룹 객체 필터링
grouped_filter = grouped.filter(lambda x: x.age.mean() < 30)
print(grouped_filter.head())
```

```python
agg_grouped = grouped.apply(lambda x: x.describe())
print(agg_grouped)
```
