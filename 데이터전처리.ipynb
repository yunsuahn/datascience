```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
print(df.head(7))
print(df.info())
```

```python
nan_deck = df['deck'].value_counts(dropna=False)
print(nan_deck)
```

```python
print(df.head().isnull())#누락데이터 찾기
print(df.isnull().sum(axis=0))#누락데이터 갯수 찾기
```

```python
missing_df = df.isnull()
for col in missing_df.columns:
  missing_count = missing_df[col].value_counts()
  try:
    print(col, ':', missing_count[True])
  except:
    print(col, ':', 0)
```

```python
df_age = df.dropna(subset=['age'], how='any', axis=0)
print(len(df_age))
```

```python
mean_age = df['age'].mean(axis=0)
df['age'].fillna(mean_age, inplace = True)#평균으로 대체, 딴거로도 대체 가능함.
```

```python
most_freq = df['embark_town'].value_counts(dropna=True).idxmax()
print(most_freq)
df['embark_town'].fillna(most_freq, inplace = True)
#df['embark_town'].fillna(method='ffill', inplace=True)
```

```python
#누락데이터가 NaN으로 안될때, NaN으로 치환 ex)?일때
df.replace('?', np.nan, inplace=True)
```

```python
df = pd.read_csv('./auto-mpg.csv', header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
print(df.head())
mpg_to_kpl = 1.60934/3.78541
df['kpl'] = df['mpg']*mpg_to_kpl
df['kpl'] = df['kpl'].round(2)#둘째자리에서 반올림
df.drop(['mpg'],axis=1,inplace=True)
print(df.head())
```

```python
print(df.dtypes)
print(df['horsepower'].unique())
df['horsepower'].replace('?',np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df['horsepower'] = df['horsepower'].astype('float')
print(df['horsepower'].dtypes)
```

```python
print(df['origin'].unique())
df['origin'].replace({1:'USA',2:'EU',3:'JPN'},inplace=True)
print(df['origin'].unique())
print(df['origin'].dtypes)
df['origin'] = df['origin'].astype('category')
print(df['origin'].dtypes)
```

```python
#데이터 구간별 분할
count, bin_dividers = np.histogram(df['horsepower'],bins=3)#3개로 분할
print(bin_dividers)
bin_names = ['low','moderate','high']
df['hp_bin'] = pd.cut(x = df['horsepower'],#데이터배열
                      bins = bin_dividers,#경계값 리스트
                      labels = bin_names,#bin이름
                      include_lowest=True)#첫 경계값 포함
print(df[['horsepower','hp_bin']].head(15))
```

```python
#더미변수 사용
horsepower_dummies = pd.get_dummies(df['hp_bin'])
print(horsepower_dummies.head(15))
```

```python
#정규화 -1~1 or 0~1
print(df.horsepower.describe())

df.horsepower = df.horsepower/abs(df.horsepower.max())
print(df.horsepower.head())
print(df.horsepower.describe())
```

```python
df = pd.read_csv('stock-data.csv')
print(df.head())
print(df.info())
```

```python
df['new_Date'] = pd.to_datetime(df['Date'])
print(df.head())
print(type(df['new_Date'][0]))
```

```python
#df.set_index('new_Date',inplace=True)
df.drop('Date', axis=1, inplace=True)
print(df.head())
```

```python
dates = ['2019-01-01', '2020-03-01', '2021-06-01']
ts_dates = pd.to_datetime(dates)

pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='A')
print(pr_year)
```

```python
#timestamp array
ts_ms = pd.date_range(start='2019-01-01',
                      end=None,
                      periods=6,
                      freq='MS',
                      tz='Asia/Seoul')
print(ts_ms)
```

```python
#period array
pr_m = pd.period_range(start='2019-01-01',
                       end=None,
                       periods=3,
                       freq='M')
print(pr_m)
```
